```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('UserCostPrediction').getOrCreate()
```

```
from pyspark.ml.regression import LinearRegression
```

```
data = spark.read.table('ecommerce_customer')
```

```
from pyspark.ml.linalg import Vectors
```

```
from pyspark.ml.feature import VectorAssembler
```

```
assembler = VectorAssembler(inputCols=['Avg Session Length','Time on App','Time on Website','Length of Membership'],outputCol='features')
```

```
output = assembler.transform(data)
```

```
output.select('features','Yearly Amount Spent').show()
```

```
final_data = output.select('features','Yearly Amount Spent')
```

```
train_data,test_data = final_data.randomSplit([0.7,0.3])
```

```
train_data.describe().show()
```

```
lr = LinearRegression(labelCol='Yearly Amount Spent',predictionCol='Predicted Amount')
```

```
lr_model = lr.fit(train_data)
```

```
test_results = lr_model.evaluate(test_data)
```

```
test_results.rootMeanSquaredError
```

```
test_results.r2
```

```
unlabelled_data = test_data.select('features')
```

```
predictions = lr_model.transform(unlabelled_data)
```

```
predictions.show()
```
